```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
 import yfinance as yf
df_google = yf.download("GOOGL")
df_google.head()
```

```python
df_google.shape
```

```python
df_google.tail()
```

## Simple Moving Average

```python
df_google['Open: 10 days rolling'] = df_google['Open'].rolling(window=10,min_periods=1).mean()
```

```python
df_google
```

```python
colors = ['green','blue']
df_google[['Open','Open: 10 days rolling']].plot(figsize=(20,10), color = colors)
```

## Cumulative Moving Average

```python
df_google['Open: CMA'] = df_google['Open'].expanding().mean()
```

```python
colors = ['green','blue']
df_google[['Open','Open: CMA']].plot(figsize=(20,10), color = colors)
```

## Exponential Moving Average

```python
df_google['Open: EMA 10 days'] = df_google['Open'].ewm(span=10).mean()
```

```python
colors = ['green','blue']
df_google[['Open','Open: EMA 10 days']].plot(figsize=(20,10), color = colors)
```

## Exponential Weighted Moving Average

```python
df_google['Open: EWMA 10 days'] = df_google['Open'].ewm(span=10).mean()
```

```python
colors = ['green','blue']
df_google[['Open','Open: EWMA 10 days']].plot(figsize=(20,10), color = colors)
```
